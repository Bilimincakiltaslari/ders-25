```python
# ana ekran oyun zemini
import time
import turtle
from turtle import Turtle
from random import randint
fon = turtle.Screen()
fon.title("TURTLE RACE")
fon.bgcolor("forestgreen")
turtle.color("white")
# YAZI VE BAŞLIK
turtle.speed(0)
turtle.penup()
turtle.setpos(-200, 200)
turtle.write("BİLİMİN ÇAKIL TAŞLARI", font=("arial",30,"bold"))
turtle.penup()
# OYUN ZEMİNİ 2. KISIM
turtle.setpos(-400, -180)
turtle.color("chocolate")
turtle.begin_fill()
turtle.pendown()
turtle.forward(800)
turtle.right(90)
turtle.forward(300)
turtle.right(90)
turtle.forward(800)
turtle.right(90)
turtle.forward(300)
turtle.end_fill()
# BİTİŞ BÖLÜMÜ
stamp_size = 20
square_size = 15
finish_line = 200
turtle.color("black")
turtle.shape("square")
turtle.shapesize(square_size / stamp_size)
turtle.penup()
for i in range(10):
    turtle.setpos(finish_line, (150-(i*square_size*2)))
    turtle.stamp()
    
for j in range(10):
    turtle.setpos(finish_line + square_size, ((150-square_size)-(j*square_size*2)))
    turtle.stamp()
    
turtle.hideturtle()
# 1. KAPLUMBAĞ
turtle1 = Turtle()
turtle1.speed(0)
turtle1.color("black")
turtle1.shape("turtle")
turtle1.penup()
turtle1.goto(-250, 100)
turtle1.pendown()
# 2. KAMLUMBAĞ
turtle2 = Turtle()
turtle2.speed(0)
turtle2.color("cyan")
turtle2.shape("turtle")
turtle2.penup()
turtle2.goto(-250, 50)
turtle2.pendown()
# 3. KAMLUMBAĞ
turtle3 = Turtle()
turtle3.speed(0)
turtle3.color("magenta")
turtle3.shape("turtle")
turtle3.penup()
turtle3.goto(-250, 0)
turtle3.pendown()
# 4. KAMLUMBAĞ
turtle4 = Turtle()
turtle4.speed(0)
turtle4.color("yellow")
turtle4.shape("turtle")
turtle4.penup()
turtle4.goto(-250, -50)
turtle4.pendown()
time.sleep(1)
pa=0
pb=0
pc=0
pd=0

for i in range(145):
    a=randint(1,5)
    b=randint(1,5)
    c=randint(1,5)
    d=randint(1,5)
    turtle1.forward(a)
    turtle2.forward(b)
    turtle3.forward(c)
    turtle4.forward(d)
    pa = pa + a
    pb = pb + b
    pc = pc + c
    pd = pd + d
ortalama = int((pa+pb+pc+pd)/4)
xa = pa - ortalama
xb = pb - ortalama
xc = pc - ortalama
xd = pd - ortalama

liste=[[pa,"siyah"],[pb,"mavi"],[pc,"mor"],[pd,"sarı"]]
liste.sort()
turtle.speed(0)
turtle.penup()
turtle.color("black")
turtle.setpos(-300, -204)
turtle.write(liste[3], font=("arial",15,"bold"))
turtle.setpos(-300, -228)
turtle.write(liste[2], font=("arial",15,"bold"))
turtle.setpos(-300, -252)
turtle.write(liste[1], font=("arial",15,"bold"))
turtle.setpos(-300, -276)
turtle.write(liste[0], font=("arial",15,"bold"))
time.sleep(3)

# 1. KAPLUMBAĞ
turtle1.goto(230-xa, 90)
turtle1.pendown()
turtle1.right(180)

# 2. KAMLUMBAĞ
turtle2.goto(230-xb, 40)
turtle2.pendown()
turtle2.right(180)

# 3. KAMLUMBAĞ
turtle3.goto(230-xc, -10)
turtle3.pendown()
turtle3.right(180)

# 4. KAPLUMBAĞ

turtle4.goto(230-xd, -60)
turtle4.pendown()
turtle4.right(180)

time.sleep(1)
for i in range(155):
    a=randint(1,5)
    b=randint(1,5)
    c=randint(1,5)
    d=randint(1,5)
    turtle1.forward(a)
    turtle2.forward(b)
    turtle3.forward(c)
    turtle4.forward(d)
    pa = pa + a
    pb = pb + b
    pc = pc + c
    pd = pd + d    

    
liste2=[[pa,"siyah"],[pb,"mavi"],[pc,"mor"],[pd,"sarı"]]
liste2.sort()
turtle.speed(0)
turtle.penup()
turtle.color("black")
turtle.setpos(-100, -204)
turtle.write(liste2[3], font=("arial",15,"bold"))
turtle.setpos(-100, -228)
turtle.write(liste2[2], font=("arial",15,"bold"))
turtle.setpos(-100, -252)
turtle.write(liste2[1], font=("arial",15,"bold"))
turtle.setpos(-100, -276)
turtle.write(liste2[0], font=("arial",15,"bold"))  
turtle.exitonclick()
turtle.mainloop()
turtle.done()
```
